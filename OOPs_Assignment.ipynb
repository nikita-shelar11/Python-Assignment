```python
#8. Write a Python program to create a class representing a shopping cart. 
#Include methods for adding and removing items, and calculating the total price. 

class Shoppingcart:
    def __init__(self):
        self.cart = {}
        self.prices = {}

    def add_items(self, item, price, quantity):
        if item in self.cart:
            self.cart[item] += quantity
        else:
            self.cart[item] = quantity
            self.prices[item] = price

    def remove_items(self, item):
        if item in self.cart:
            del self.cart[item]
            del self.prices[item]
        else:
            print("Item not found")

    def view_cart(self):
        print("View cart items:")
        for item in self.cart:
            qty = self.cart[item]
            price = self.prices[item]
            print(f"{item} - Quantity: {qty}, Price: {price}, Total: {qty*price}")

    def total_price(self):
        total_price = 0
        for item in self.cart:
            total_price += self.cart[item] * self.prices[item]
        return total_price

SC = Shoppingcart()
SC.add_items("Shoes", 3000, 2)
SC.add_items("Shirt", 2000, 3)
SC.add_items("Watch", 5000, 1)
SC.add_items("Dress", 3500, 4)

SC.view_cart()

print("Total checkout price is:",SC.total_price())

SC.remove_items("Watch")

SC.view_cart()

print("Total checkout price is:",SC.total_price())
```

```python
#12. Build a flashcard using class in python. A flashcard is a card having information on both sides, which can be used as an aid in memoization. 
#Flashcards usually have a question on one side and an answer on the other.
'''
Example 1:

Approach:

Create a class named FlashCard.
Initialize dictionary fruits using init() method. Here you have to define fruit name as key and it's color as value.
E.g., {"Banana": "yellow", "Strawberries": "pink"}
Now randomly choose a pair from fruits by using random module and store the key in variable fruit and value in variable color.
Now prompt the user to answer the color of the randomly chosen fruit.
If correct print correct else print wrong.
'''

import random

class FlashCards:
    def __init__(self):
        self.fruits = {
            "Banana" : "yellow",
            "Strawberry" : "pink",
            "Watermelon" : "green",
            "Orange" : "orange",
        }

    def quiz(self):
        fruit, color = random.choice(list(self.fruits.items()))

        user_answer = input(f"What is the color of the {fruit}?")
        
        if user_answer.lower() == color.lower():
            print("Correct answer!")
        else:
            print("Wrong answer...")

play = FlashCards()

while True:
    play.quiz()

    choice = int(input("Press 1 to play the quiz again or press 0 to exit"))

    if choice == 0:
        print("Quiz over. good game!!!")
        break
```

```python
#14. Write a program that uses datetime module within a class. Enter manufacturing date and expiry date of the product. 
#The program must display the years, months and days that are left for expiry.

from datetime import datetime

class Product:
    def __init__(self, mfg_date, exp_date):
        self.mfg_date = datetime.strptime(mfg_date, "%d-%m-%Y")
        self.exp_date = datetime.strptime(exp_date, "%d-%m-%Y")

    def time_remain(self):
        today = datetime.now()

        if self.exp_date < today:
            print("Product expired!")

        diff = self.exp_date - today
    
        days = diff.days
        years = days // 365
        months = (days % 365) // 30
        remaining_days = (days % 365) % 30
    
        print("Time left for expiry:")
        print("Years:", years)
        print("Months:", months)
        print("Days:", remaining_days)

mfg = input("Enter manufacturing date (dd-mm-yyyy): ")
exp = input("Enter expiry date (dd-mm-yyyy): ")

p = Product(mfg, exp)
p.time_remain()
```

```python
#19. Write a Python class Restaurant with attributes like menu_items, book_table, and customer_orders, and 
#methods like add_item_to_menu, book_tables, and customer_order.
'''
	Perform the following tasks now:

	Now add items to the menu.
	Make table reservations.
	Take customer orders.
	Print the menu.
	Print table reservations.
	Print customer orders.
	Note: Use dictionaries and lists to store the data.
'''

class Restaurant:
    def __init__(self):
        self.menu_items = []
        self.book_table = {}
        self.customer_orders = {}

    def add_item_to_menu(self, item_name, price):
        item = {"name": item_name, "price": price}
        self.menu_items.append(item)
        print(f"{item_name} is added to the menu.")

    def book_tables(self, table_no, customer_name):
        if table_no in self.book_table:
            print(f"Table {table_no} is already booked!")
        else:
            self.book_table[table_no] = customer_name
            print(f"Table {table_no} is booked for {customer_name}")

    def customer_order(self, customer_name, item):
        self.customer_orders[customer_name] = item
        print(f"Order taken for {customer_name}")

    def print_menu(self):
        print("Menu")
        for item in self.menu_items:
             print(f"{item['name']} - Rs {item['price']}")

    def print_table_reservations(self):
        print("Table Reservations:")
        for table, name in self.book_table.items():
            print(f"Table {table} -> {name}")

    def print_customer_orders(self):
        print("Customer Orders:")
        for customer, items in self.customer_orders.items():
            print(f"{customer} ordered: {items}")
        
r = Restaurant()
r.add_item_to_menu("Pasta",100)
r.add_item_to_menu("Noodles", 50)
r.add_item_to_menu("Pizza", 200)
r.add_item_to_menu("Burger", 250)

r.book_tables(1, "Ada")
r.book_tables(1, "Bliss")

r.customer_order("Ada", "Pasta")
r.customer_order("Bliss", "Pizza")

print(r.print_menu())

r.print_table_reservations()

r.print_customer_orders()
```

```python
#5. Write a Python program to create a class representing a binary search tree. 
#Include methods for inserting and searching for elements in the binary tree.

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert_recursive(self.root, value)

    def _insert_recursive(self, current, value):
        if value < current.value:
            if current.left is None:
                current.left = Node(value)
            else:
                self._insert_recursive(current.left, value)

        elif value > current.value:
            if current.right is None:
                current.right = Node(value)
            else:
                self._insert_recursive(current.right, value)

        else:
            print(f"{value} already exists in the tree.")

    def search(self, value):
        return self._search_recursive(self.root, value)

    def _search_recursive(self, current, value):
        if current is None:
            return False

        if current.value == value:
            return True

        if value < current.value:
            return self._search_recursive(current.left, value)
        else:
            return self._search_recursive(current.right, value)

bst = BinarySearchTree()

bst.insert(70)
bst.insert(50)
bst.insert(80)
bst.insert(40)
bst.insert(90)
bst.insert(60)
bst.insert(100)

print(bst.search(40)) 
print(bst.search(10)) 
```

```python
#6. Write a Python program to create a class representing a stack data structure. Include methods for pushing and popping elements.

class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)
        print(f"{item} pushed to stack.")

    def pop(self):
        if len(self.stack) == 0:
            print("Stack is empty! Cannot pop.")
            return None
        return self.stack.pop()

    def display(self):
        print("Current Stack:", self.stack)

s = Stack()

s.push(10)
s.push(20)
s.push(30)

s.display()

print("Popped:", s.pop())
print("Popped:", s.pop())

s.display()
```

```python
#7. Write a Python program to create a class representing a linked list data structure. 
#Include methods for displaying linked list data, inserting and deleting nodes.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            print(f"{data} inserted as head node.")
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        print(f"{data} inserted.")

    def delete(self, key):
        temp = self.head

        if temp is not None and temp.data == key:
            self.head = temp.next
            print(f"{key} deleted from list.")
            return

        prev = None
        while temp is not None and temp.data != key:
            prev = temp
            temp = temp.next

        if temp is None:
            print(f"{key} not found in list.")
            return

        prev.next = temp.next
        print(f"{key} deleted from list.")

    def display(self):
        if self.head is None:
            print("Linked List is empty.")
            return

        temp = self.head
        print("Linked List:")
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

ll = LinkedList()

ll.insert(10)
ll.insert(20)
ll.insert(30)
ll.insert(40)

ll.display()

ll.delete(20)
ll.delete(100) 

ll.display()
```

```python
#10. Write a Python program to create a class representing a queue data structure. Include methods for enqueueing and dequeueing elements.

class Queue:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)
        print(f"{item} added to queue.")

    def dequeue(self):
        if len(self.queue) == 0:
            print("Queue is empty! Cannot dequeue.")
            return None
        removed = self.queue.pop(0)
        print(f"{removed} removed from queue.")
        return removed

    def display(self):
        print("Current Queue:", self.queue)

q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

q.display()

q.dequeue()
q.dequeue()

q.display()
```
