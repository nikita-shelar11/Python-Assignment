```python
#4. Running Sum on list Write a program to print a list after performing running sum on it.

list1 = [1,2,3,4,5,6]

Running_sum = []
total = 0
for num in list1:
    total += num
    Running_sum.append(total)

print("Running sum:",Running_sum)
```

```python
#5. You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

nums = [2,4,6,10,1]
result = []

for i in range(len(nums)):
    total = 0
    for j in range(len(nums)):
        if nums[j] >= nums[i]:
            total += nums[j]
    result.append(total)

print(result)
```

```python
#6. Find list of common unique items from two list. and show in increasing order

num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]

common_items = []

for i in num1:
    if i in num2 and i not in common_items:
        common_items.append(i)
        
common_items.sort()
print("Common unique items:",common_items)
```

```python
#9. Shortlist Students for a Job role. Ask user to input students record and store in tuples for each record. 
#Then ask user to input three things he wants in the candidate- Primary Skill, Higher Education, Year of Graduation.

Student_records = []

n = int(input("Enter number of records:"))

for i in range(n):
    print(f"Enter details of student: {i+1}")
    
    name = input("Enter student name:")
    education = input("Enter higher education:")
    skill = input("Enter primary skill:")
    year = int(input("Enter year of graduation:"))

    Student_records.append((name, education, skill, year))
    
print(f"\nEnter Job Role Requirement")
req_skill = input("Enter Skill:")
req_education = input("Enter Higher Education:")
req_year = int(input("Enter Year of Graduation:"))

for student in Student_records:
    name, education, skill, year = student
    
    if(
        req_skill == skill and
        req_education == education and
        req_year == year
    ):
        print("Candidate found:",student)
        break
else:
    print("No such candidate found")
```

```python
#14. Write a Python function that takes a list and returns a new list with unique elements of the first list.

#lst = [1,2,3,3,3,3,4,5]

def distinct_list(lst):
    unique_list = []
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
```

```python
print(distinct_list([1,2,3,3,3,3,4,5]))
```

```python
#16. Write a Python function to check whether a number is perfect or not.

def perfect_num(n):
    divisor_sum = 0
    for i in range(1, n):
        if n % i == 0:
            divisor_sum += i
    return divisor_sum == n
```

```python
perfect_num(8128)
```

```python
#18. Write a python function that receives a list of integers and prints out a histogram of bin size 10

def histogram(data):
    bins = {}

    for num in data:
        key = (num // 10) * 10 + 1   # 11, 21, 31...
        label = f"{key}-{key+9}"
        bins[label] = bins.get(label, 0) + 1

    print(bins)


# Example
nums = [13,42,15,37,22,39,41,50]
histogram(nums)
```
