```python
#2. Write a Python program to read first n lines of a file.

n = int(input("Enter number of lines:"))

with open("demo.txt", "r") as f:
    lines = f.readlines()
    for line in lines[:n]:
        print(line, end="")      
```

```python
#3. Write a Python program to append text to a file and display the text.

with open("demo.txt", "a") as f:
    text = f.write("\nI also want to learn AI, ML.")
print(text)
```

```python
#4. Write a Python program to read last n lines of a file.

n = int(input("Enter number of lines:"))

with open("demo.txt", "r") as f:
    lines = f.readlines()
    for line in lines[-n:]:
        print(line, end="")
```

```python
#5. Write a Python program to read a file line by line and store it into a list.
list_demo = []

with open("demo.txt","r") as f:
    for line in f:
        list_demo.append(line.strip())
print(list_demo)
```

```python
#6. Write a Python program to read a file line by line store it into a variable.

list = []

with open("demo.txt","r") as f:
    for line in f:
        list.append(line.strip())
print("List:",list)
```

```python
#7. Write a python program to find the longest words.

with open("demo.txt","r") as file:
    words = file.read().split()
    max_len = max(len(word) for word in words)
    for word in words:
        if len(word) == max_len:
            longest_word = word
            print("Longest word:",longest_word)
```

```python
#8. Write a Python program to count the number of lines in a text file.

with open("demo.txt", "r") as file:
    content = file.readlines()
    count_lines = len(content)
    print("Number of lines:",count_lines)
```

```python
#9. Write a Python program to count the frequency of words in a file.

with open("demo.txt", "r") as file:
    words = file.read().split()

freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
print(freq)
```

```python
#10. Write a Python program to get the file size of a plain file.

import os

file = "demo.txt"
size = os.path.getsize(file)
print("File size:",size,"bytes")
```

```python
#11. Write a Python program to write a list to a file.

list = ["India","USA","Nepal","Saudi","Iran"]

with open("demo.txt","w") as f:
    for item in list:
        f.write(item + "\n")
```

```python
#12. Write a Python program to copy the contents of a file to another file .

with open("demo.txt","r") as file:
    data = file.read()

with open("new_demo.txt", "w") as copy_file:
    copy_file.write(data)
```

```python
#13. Write a Python program to read a random line from a file

import random

with open("demo.txt","r") as file:
    lines = file.readlines()
    read_random = random.choice(lines)
print("Random line: ",read_random)
```

```python
#14. Write a Python program to assess if a file is closed or not.

with open("demo.txt","r") as file:
    print("File is closed:",file.closed)

print("File is not closed:",file.closed)
```
